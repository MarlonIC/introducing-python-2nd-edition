#### Obtener todas las llaves con keys()

```python
signals = {
    'green': 'go',
    'yellow': 'go faster',
    'red': 'smile for the camera'
}
```

```python
signals.keys()
```

```python
list(signals.keys())
```

#### Obtener todos los valores con values()

```python
signals.values()
```

```python
list(signals.values())
```

#### Obtener todas las llaves y valores con items()

```python
signals.items()
```

```python
list(signals.items())
```

#### Obtener su longitud con len()

```python
len(signals.items())
```

#### Combinar diccionarios

```python
first = {'a': 'agony', 'b': 'bliss'}
```

```python
second = {'b': 'bagels', 'c': 'candy'}
```

```python
{**first, **second}
```

```python
third = {'d': 'donuts'}
```

```python
{**first, **second, **third}
```
